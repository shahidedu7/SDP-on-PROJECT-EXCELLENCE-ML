```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
from sklearn.cluster import AgglomerativeClustering
```

```python
customer_data = pd.read_csv('customer.csv')
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data.shape
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendrograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
```

```python
labels=cluster.fit_predict(data)
```

```python
labels
```

```python
clusters=pd.DataFrame(labels,columns=['Clusters'])
```

```python
clusters
```

```python
data=pd.concat([customer_data,clusters],axis=1)
```

```python
data
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=labels, cmap='rainbow')
```

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
Univ
```

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:])
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, metric = 'euclidean', linkage = 'single')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
df_norm['h_clusterid'] = y_hc
```
