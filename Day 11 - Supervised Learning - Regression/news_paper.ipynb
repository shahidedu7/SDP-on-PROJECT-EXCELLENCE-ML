# Fitting a Linear Regression Model

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
X = df[["daily"]]
y = df["sunday"]
```

```python
X
```

```python
y
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.intercept_      # y = a + b X   a is intercept, b is coeffi
```

```python
model.coef_
```

```python
# sunday = model.intercept_ + model.coef_* daily
```

```python
y
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot.scatter(x="daily",y="sunday")
plt.plot(df["daily"],y_pred,color="red")
```

```python
# R^2 (coefficient of determination of the prediction)
model.score(X,y)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y, y_pred)
mae
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, y_pred)
mse
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y, y_pred)
rmse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y, y_pred)
r2
```
